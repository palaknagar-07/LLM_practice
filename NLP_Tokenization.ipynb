```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
text = "Hello my name is Palak Nagar. I am very good in python and DSA. How are you?"
sentences = sent_tokenize(text)
print(sentences)



from nltk.tokenize import word_tokenize
text1 = "Dr. Smith said he doesn't like it."
sentence2 = word_tokenize(text1)
print(sentence2)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize

text = "Hello!! how are you?"

sent_token = sent_tokenize(text)
word_token = word_tokenize(text)
punc_token = wordpunct_tokenize(text)

print(sent_token)
print(word_token)
print(punc_token)
```

```python
import re


text = "Email john.doe@example.com and visit https://example.com"

# Custom pattern to preserve emails and URLs
pattern = r'https?://\S+|[\w.-]+@[\w.-]+|[\w]+'
tokens = re.findall(pattern, text)
# ["Email", "john.doe@example.com", "and", "visit", "https://example.com"]
print(tokens)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Dr. Smith sent emails to john@example.com yesterday."
doc = nlp(text)

for token in doc:
    print(token.text)
```

```python
from  transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
# text = " Dr. Smith sent emails to john@example.com yesterday."
text = "unsentimental"
tokens = tokenizer.tokenize(text)

tokens_to_ids = tokenizer.convert_tokens_to_ids(tokens)
print(tokens)
print(tokens_to_ids)
```
